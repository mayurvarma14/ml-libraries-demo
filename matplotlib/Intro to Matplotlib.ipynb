### Different methods to plot graph

**Method 1**

```python
fig=plt.figure() #creates a figure
ax=fig.add_subplot() #adds some axes
plt.show()
```

**Method 2**

```python
fig=plt.figure() #creates a figure
ax=fig.add_subplot() #adds some axes
ax.plot(x,y) #add data
plt.show()
```

**Method 3 (recommended)**

```python
fig, ax=plt.subplots()
ax.plot(x,[50,100,200,250]);
```
